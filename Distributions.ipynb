### Chi-Square Distribution

Distribution of a sum of squares $X_1^{2}, X_2^{2}, ..., X_n^{2}$ sampled from the distribution $N(0, 1)$ has gamma distribution with parameters $\alpha=\frac{n}{2}, \lambda=\frac{1}{2}$. The proof is based on the summation property of the gamma distribution. Chi-square distribution is a distribution of a variable which is a sum of squares 

$Y_n = \sum_{i=0}^n X_i^{2}$            


$X_i$ ~ $N(0, 1)$

The expected value $EX = n$ and variance $VarX = 2n$

Here we compare empirical distribution, distribution computed based on gamma distribution and the actual chi-square

```python
from scipy.stats import norm, chi2, gamma

def sum_of_squares(n, std=1):
    return np.sum(norm.rvs(size=n, scale=std) ** 2)

def generate(m, n, std=1):
    return sorted([sum_of_squares(n, std=std) for _ in range(m)])

df = 10
size = 200

fig, ax = plt.subplots(2)

y = generate(2000, df)
ax[0].hist(y, density=True, histtype='step', bins='auto')

''''''
l1, = ax[1].plot(y, gamma.pdf(y, a=df / 2, scale=2), color='orange')
l2, = ax[1].plot(y, chi2.pdf(y, df), color='blue')

plt.legend([l1, l2], ['empirical', 'theoretical'])
```

Quantile of order p is any number for which the following are true:

$F(X < x_p) >= p$

$F(X > x_p) >= 1 - p$

As df increases, we can use normal distribution quantiles to approximate quantiles of Chi-Square distribution

```python
def compute_quantile(p, df):
    return ((np.sqrt(2 * df - 1) + norm.ppf(p)) ** 2) / 2

print(chi2.ppf(0.8, 50))
print(compute_quantile(0.8, 50))
```

A manufacturer of car batteries claims that the life of the company’s batteries is
approximately normally distributed with a standard deviation equal to 0.9 year.
If a random sample of 10 of these batteries has a standard deviation of 1.2 years,
do you think that σ > 0.9 year? Use a 0.05 level of significance.

We use a test statistic 

$$ U = \frac{nS^2}{\sigma_0^2} $$ which has \$ \chi^2 \$ distribution with \$ n - 1 \$ degrees of freedom

$$ H_0 : \sigma^2 = 0.9 $$
$$ H_1 : \sigma^2 > 0.9 $$

```python
fig, ax = plt.subplots(figsize=(7, 5))

alpha = 0.05
n = 10

# this would be for the two-sided test
q = chi2.ppf(1 - alpha, n - 1)

print(f"Critical region : <{q}, inf>")

l10 = plot_chi2(n, ax)
ax.axvline(q, c='r')

std = 1.2
h_std = 0.9
# now, depending on whether we multiply by n or (n - 1) we get different 
# results, but in most coursebooks the factor is n
stat = n * std ** 2 / h_std ** 2
print(stat > q)
```

### The Distribution of the Statistics (mean(X), S)

Let statistic X have a normal distribution $N(0, \sigma)$. We draw a sample of size n and compute the mean $\mu$ and the standard deviation s, 

$\overline x = \frac{1}{n} \sum_{k=1}^{n} x_k$

$s^{2} = \frac{1}{n} \sum_{k=1}^{n} x_k^{2} - \overline x^{2}$

It is obvious that for a different samples values of $\overline x$ and $s$ will vary, so they are also a random variables

$\overline X = \frac{1}{n} \sum_{k=1}^{n} X_k$

$S^{2} = \frac{1}{n} \sum_{k=1}^{n} X_k^{2} - \overline X^{2}$

where $\overline X$ has a normal distribution $N(\mu, \frac{\sigma}{\sqrt{n}})$ and $S^{2}$ has chi-square distribution of n - 1 degrees of freedom

It can be proved that the join distribution of $\overline X$ and $S^{2}$ is a product of the distributions so that $\overline X$ and $S^{2}$ are independent. It is a very important result and practically it means that if we sample from a population and group samples based on similar, say, $s^{2}$, then for each group the distribution of $\overline x$ will be approximately the same (and vice-versa). 

An example: 
We draw a sample from the normal distribution ~ N(1, 2). We are interested in finding out how likely it is to obtain a standard deviation greater then the one computed from the drawn sample. 
We compute the mean $\overline x$ and standard deviation $s^{2}$. Z variable is equal to $z = n s^{2}$ and we want to answer a question 
$P(Z >= z) = P(\frac{nS^{2}}{\sigma^2} >= \frac{z}{4}) = P(\chi^2 >= \frac{z}{4})$

We use the $\chi^2$ distribution with 11 degrees of freedom

```python
# We draw a sample from the normal distribution ~ N(1, 2)
x = np.array([2.0, 2.5, 0.5, 1.0, 0.0, -0.9, 5.1, -1.5, 0.8, 1.1, 0.8, 0.4])
mean = np.mean(x)
s_2 = np.var(x)
print(f"mean = {mean}")
print(f"s^2 = {s_2}")
z = len(x) * s_2
print(f"z = {z}")
p = 1 - chi2.cdf(z / 4, df=len(x) - 1)
print(f"p = {p}")
```

### Student's t-Distribution 

Consider the statistics $\overline X$ and $S^{2}$ (given in the section above) where $X_1, X_2, ..., X_n$ are independent and have the same normal distribution $N(\mu, \sigma)$. Then $\overline X$ has the normal distribution $N(\mu, \frac{\sigma}{\sqrt{n}})$. Hence, for a known $\mu$ and unknown $\sigma$, the distribution of $\overline X$ is unknown. We cannot approximate 
$\sigma$ by the S statistics since it is itself a random variable. So in order to be able to deduce anything about $\mu$ without the knowledge of $\sigma$ we need a statistic which is a function of $\mu$ and with a distribution independent of $\sigma$. The solution to the problem is Student's t-statistics.

t value is defined as 

$t = \frac{\overline X - \mu}{S}\sqrt{n - 1}$

The distribution is symmetric but has fatter tails than the normal distribution, (hence the excess should be > 0) and as the $df -> \infty$, the distribution resembles the normal distribution.

For df=1 we have Cauchy distribution

```python
from scipy.stats import t 

fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True)

x = np.linspace(-5, 5, 1000)

l1, = ax.plot(x, t.pdf(x, df=1))
l2, = ax.plot(x, t.pdf(x, df=2))
l5, = ax.plot(x, t.pdf(x, df=5))
l20, = ax.plot(x, t.pdf(x, df=20))

ln, = ax.plot(x, norm.pdf(x))

plt.legend([l1, l2, l5, l20, ln], ['df=1', 'df=2', 'df=5', 'df=20', 'norm'])
```

Kurtosis comparison - it gives information on how fat the tails are. It has nothing to do with the peakedness! Here kurtosis is really the excess since it is computed in relation to the normal distribution value, where kurtosis = 3

```python
from scipy.stats import uniform, t
df = 20
mean, var, skew, kurt = t.stats(df, moments='mvsk')
print(f"mean = {mean}, var = {var}, skew = {skew}, kurt = {kurt}")
mean, var, skew, kurt = norm.stats(0, 1, moments='mvsk')
print(f"mean = {mean}, var = {var}, skew = {skew}, kurt = {kurt}")
mean, var, skew, kurt = uniform.stats(-1, 2, moments='mvsk')
print(f"mean = {mean}, var = {var}, skew = {skew}, kurt = {kurt}")

x = np.linspace(-3, 3, 1000)
fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True)
ax.plot(x, t.pdf(x, df), label="t")
ax.plot(x, norm.pdf(x, 0, 1), label="norm")
ax.plot(x, uniform.pdf(x, -1, 2), label="uniform")
```

### Bernoulli Distribution

It models the number of successes in a trial of n independent experiments. 

We can see that for p > 0.5 it is left-skewed, for p < 0.5 it is right-skewed and for p = 0.5 it is symmetric

```python
from scipy.stats import binom

n = 10
x = np.arange(0, n + 1)
plt.plot(x, binom.pmf(x, n, 0.8), marker='.', label='p=0.8')
plt.plot(x, binom.pmf(x, n, 0.5), marker='.', label='p=0.5')
plt.plot(x, binom.pmf(x, n, 0.2), marker='.', label='p=0.2')
plt.legend(loc=1)
```

Let's model how well Poisson distribution models Bernoulli distribution.
Poisson approximation should be used when the number of trials is large (n >= 20) and the
probability of success is small (p <= 0.05)

```python
from scipy.stats import poisson

def plot_compare(ax, n, p):
    x = np.arange(0, n + 1)
    lmbd = n * p
    ax.plot(x, binom.pmf(x, n, p), marker='.', label=f"n={n}, p={p}")
    ax.plot(x, poisson.pmf(x, lmbd), marker='.', label=f"lmbd={2}")
    ax.legend()
    

fig, axes = plt.subplots(4, figsize=(10, 10))
plot_compare(axes[0], 4, 0.5)
plot_compare(axes[1], 10, 0.2)
plot_compare(axes[2], 20, 0.1)
plot_compare(axes[3], 200, 0.01)
```

### Moivre Laplace Theorem

If the number of trials is large enough, we can approximate the number of successes with the probability density function of the normal distribution, that is

$$ P(K = k) \approx f_Z(k) $$ where 
\$ Z \sim N(np, npq)\$

```python
from scipy.stats import norm

def plot_compare(ax, n, p):
    x = np.arange(0, n + 1)
    lmbd = n * p
    std = np.sqrt(n * p * (1 - p))
    ax.plot(x, binom.pmf(x, n, p), marker='.', label=f"n={n}, p={p}")
    ax.plot(x, norm.pdf(x, lmbd, std), marker='.', label=f"norm")
    ax.legend()
    
fig, axes = plt.subplots(4, figsize=(10, 10))
plot_compare(axes[0], 4, 0.5)
plot_compare(axes[1], 10, 0.2)
plot_compare(axes[2], 20, 0.1)
plot_compare(axes[3], 200, 0.01)
```
